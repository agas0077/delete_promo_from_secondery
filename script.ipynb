```python
import pandas as pd
```

```python
salestool = pd.read_excel("SALESTOOL.xlsx", converters={'bp': str}, na_values=0)
salestool = salestool.fillna(0)
salestool
```

```python
promo = pd.read_excel("promo.xlsx")
promo
```

```python
promo = promo.dropna(axis=1)
```

```python
uniqueBP = {}
for index, row in promo.iterrows():
    bp = row["ProductCode"]
    if bp not in uniqueBP:
        uniqueBP[bp] = []
    soweeks = range(int(row["SOWeekStart"]), int(row["SOWeekEnd"]) + 1)
    for week in soweeks:
        if week not in uniqueBP[bp]:
            uniqueBP[bp].append(week)
print(uniqueBP)
```

```python
for bp, weeks in uniqueBP.items():
    weeks.sort()
```

```python
for index, row in salestool.iterrows():
    bp = row['bp']
    week = row['week']
    if bp in uniqueBP and week in uniqueBP[bp]:
        salestool = salestool.drop(index)
    row[bp] = str(bp)
```
